```python
# seeing the final 20 rows
df.tail(20)
```

ok now i can sum the number of yellow cards to do my hist plot

```python
#ok i will use hist plot to see wich teams
#got the most cards(hist will take the number of cards)
#frist i will get the sum of yellow and red cards
search='ycard'
# Check if the 'card' column is of string type before applying str accessor
if df['card'].dtype == 'object':
    df_Y = df['card'].str.count(search)
else:
    # Handle cases where the column is not string, by converting it first
    df['card'] = df['card'].astype(str)
    df_Y = df['card'].str.count(search)

print(df_Y.sum())
#now i can do a hist to see how does my data looks like
dfc('hist','number of card','number of teams','number of yellow cards(x) to teams(y)',df_Y)
```

most teams got five yellow cards

now i will see the red cards

```python
# now i will do the same if statment with red cards
search='rcard'
# Check if the 'card' column is of string
#type before applying str accessor
if df['card'].dtype == 'object':
    df_R = df['card'].str.count(search)
    print(df_R.sum())
else:
    # Handle cases where the column is not string, by converting it first
    df['card'] = df['card'].astype(str)
    df_R = df['card'].str.count(search)
    print(df_R.sum())
#now i can do a hist to see how does my data looks like
dfc('hist','number of card','number of teams','number of red cards(x) to teams(y)',df_R)
```

oh most teams got 0 red cards

ok now i will see if there is a realation

```python
# frist i will groupy the goals with yellow cards and use a plot
# then i will do this with red
# Calculate the sum of yellow and red cards for each home_team_goal group
if df['card'].dtype == 'object':
    search='ycard'
    df_Y = df['card'].str.count(search)
    search='rcard'
    df_R = df['card'].str.count(search)
else:
    #now we see if the cards are not object
     #(cards are object but i want to make sure that
    #my code works with you) we will make it str
    df['card'] = df['card'].astype(str)
    search='ycard'
    df_Y = df['card'].str.count(search)
    search='rcard'
    df_R = df['card'].str.count(search)
# lambada funiction make us sum the number of y cards
# and iloc[x.index] takes the values in the corresponding row
#to cards number
df_YY=df.groupby('home_team_goal').apply(lambda x: df_Y.iloc[x.index].sum() + df_R.iloc[x.index].sum())
dfc('bar','home_team_goal','number of cards','home team goals vs the number of cards',df_YY)
```

we could notice that the teams wich got the most goals got less cards than the teams wich got the least cards
